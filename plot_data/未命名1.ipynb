```python
import matplotlib.pyplot as plt

# empty lists
x_els, y_els = [], []
x_gamma, y_gamma = [], []
x_inl, y_inl = [], []
x_total, y_total = [], []

# els data
with open('./C/C-n-els.txt') as f:
    for line in f:
        if line.strip():
            parts = line.split()
            x_els.append(float(parts[0]))
            y_els.append(float(parts[1]))

# gamma data
with open('./C/C-n-gamma.txt') as f:
    for line in f:
        if line.strip():
            parts = line.split()
            x_gamma.append(float(parts[0]))
            y_gamma.append(float(parts[1]))

# inl data
with open('./C/C-n-inl.txt') as f:
    for line in f:
        if line.strip():
            parts = line.split()
            x_inl.append(float(parts[0]))
            y_inl.append(float(parts[1]))

# total data
with open('./C/C-n-total.txt') as f:
    for line in f:
        if line.strip():
            parts = line.split()
            x_total.append(float(parts[0]))
            y_total.append(float(parts[1]))


plt.yscale('log')
plt.xscale('log')


plt.plot(x_els, y_els, label='n, els')
plt.plot(x_gamma, y_gamma, label='n, gamma')
plt.plot(x_inl, y_inl, label='n, inl')
plt.plot(x_total, y_total, label='n, total')


plt.xlabel('Energy (MeV)')
plt.ylabel('Cross Section')
plt.title('Carbon (n,x) Reactions')
plt.legend()


# plt.show()
plt.savefig('./carbon.png')
```
